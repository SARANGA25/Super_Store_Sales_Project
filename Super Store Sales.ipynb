### Sales By Sub Category

```python
Sales_by_Sub_Cat = df.groupby(['Sub-Category'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)
ax=sns.barplot(x='Sales', y='Sub-Category', data=Sales_by_Sub_Cat, palette="viridis")
sns.set(rc={'figure.figsize':(15,10)})
                                                                                           
for bar in ax.patches:
    ax.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width()/1000:.1f}k', 
            va='center', ha='left', color='black')
   
    
plt.title("Sales By Sub Category", fontweight='bold')
plt.ylabel('Sub Category' ,fontweight='bold')
plt.xlabel('Sales' ,fontweight='bold')
plt.show()
```

## Ship Mode

```python
Ship_Mode = df.groupby(['Ship Mode'], as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)
Ship_Mode.head
```

```python
Ship_Mode = df.groupby(['Ship Mode'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)
ax = sns.barplot(x='Sales', y='Ship Mode', data=Ship_Mode, palette="viridis")

for bar in ax.patches:
    ax.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width()/1000:.1f}k', 
            va='center', ha='left', color='black')
    
sns.set(rc={'figure.figsize':(15,3)})
plt.title("Total Sales By Ship Mode", fontweight='bold')
plt.xlabel("Total Sales" ,fontweight='bold')
plt.ylabel("Ship Mode" ,fontweight='bold')
plt.show()
```

## Profit By Month

```python
Profit_by_Mon = df.groupby(['Month'], as_index=False)['Profit'].sum().sort_values(by='Profit', ascending=False)
ax = sns.barplot(x='Month', y='Profit', data=Profit_by_Mon)

for bar in ax.patches:
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height()/1000:.1f}k', 
            ha='center', va='bottom', color='black')
    
sns.set(rc={'figure.figsize':(15,3)})
plt.title("Total Profit by Month", fontweight='bold')
plt.xlabel("Month" ,fontweight='bold')
plt.ylabel("Total Profit" , fontweight='bold')
plt.show()
```

## Sales by State

```python
sales_by_state = df.groupby(['State'], as_index=False)['Sales'].sum().sort_values(by='Sales' , ascending=False).head(20)
ax=sns.barplot(x='Sales',y='State', data=sales_by_state, palette="viridis")
sns.set(rc={'figure.figsize':(15,10)})

for bar in ax.patches:
    ax.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width()/1000:.1f}k', 
            va='center', ha='left', color='black')
    
plt.title("Sales OF State", fontweight="bold")
plt.ylabel('State', fontweight="bold")
plt.xlabel('Sales', fontweight="bold")
plt.show()
  
```

### Sales By Region

```python
Sales_Region =df.groupby('Region')['Sales'].sum()
percentage_sales=(Sales_Region/Sales_Region.sum())*100
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

plt.gcf().set_size_inches(5, 5)
plt.pie(Sales_Region, labels=Sales_Region.index, autopct='%1.1f%%', startangle=100, wedgeprops=dict(width=0.4))
plt.title('Sales Distribution by Region', fontweight="bold") 
plt.show()
```

## Sales By Payment Mode

```python
sales_By_Payment = df.groupby('Payment Mode')['Sales'].sum()
percentage_sales = (sales_By_Payment / sales_By_Payment.sum()) * 100

# Define colors for each slice of the pie chart
colors = ['gold', 'lightcoral', 'lightskyblue' ]

plt.figure(figsize=(5, 5))
plt.pie(sales_By_Payment, labels=sales_By_Payment.index, autopct='%1.1f%%', startangle=100, wedgeprops={'width': 0.4}, colors=colors)
plt.title('Sales Distribution by Payment Mode', fontweight="bold")
plt.show()
```

## Sales By Segment

```python
sales_By_segment = df.groupby('Segment')['Sales'].sum()
percentage_sales = (sales_By_segment/sales_By_segment.sum())*100
colors = ['blue', 'green', 'red']  # Add more colors as needed

plt.figure(figsize=(5,5))
plt.pie(sales_By_segment, labels=sales_By_segment.index, autopct='%1.1f%%', startangle=100, wedgeprops={'width':0.4}, colors=colors)
plt.title('Sales Distribution by Segment', fontweight='bold')
plt.show()
```
